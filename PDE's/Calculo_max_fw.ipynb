```python
import sympy as sp


Sw , fw = sp.symbols('Sw fw')

fw = ((0.5 * ((Sw - 0.1) /(1 - 0.1 - 0.2))**2.0)/1.28)/(((0.5 * ((Sw - 0.1) /(1 - 0.1 - 0.2))**2.0)/1.28) + ((0.9 * (1 - ((Sw - 0.1)/(1 - 0.1 - 0.2)))**3.0)/1.22))

derivada_1 = sp.diff(fw,Sw)
derivada_2 = sp.diff(derivada_1,Sw)

ponto_maximo = sp.nsolve(derivada_2,Sw,0.2,verify = False)

valor_maximo = derivada_1.subs(Sw,ponto_maximo)


print(valor_maximo)
```
